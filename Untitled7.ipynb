# Task
Converter uma imagem colorida em formato PNG para níveis de cinza e para binarizada (preto e branco).

```python
import cv2

image = cv2.imread('/image.png')
```

```python
if image is not None:
  gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
else:
  print("Error: Could not process image. The image variable is not loaded.")
```

```python
if 'gray_image' in locals() and gray_image is not None:
  _, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
else:
  print("Error: The grayscale image is not available.")
```

```python
import matplotlib.pyplot as plt
import cv2

plt.figure(figsize=(15, 5))

if 'image' in locals() and image is not None:
  plt.subplot(1, 3, 1)
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title('Imagem Original')
  plt.axis('off')
else:
  print("Error: Original image could not be displayed.")

if 'gray_image' in locals() and gray_image is not None:
  plt.subplot(1, 3, 2)
  plt.imshow(gray_image, cmap='gray')
  plt.title('Imagem em Níveis de Cinza')
  plt.axis('off')
else:
  print("Error: Grayscale image could not be displayed.")

if 'binary_image' in locals() and binary_image is not None:
  plt.subplot(1, 3, 3)
  plt.imshow(binary_image, cmap='gray')
  plt.title('Imagem Binarizada')
  plt.axis('off')
else:
  print("Error: Binary image could not be displayed.")

plt.show()
```
